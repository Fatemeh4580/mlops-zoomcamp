```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Lasso, Ridge
import matplotlib.pyplot as plt
from sklearn.feature_extraction import DictVectorizer
from scipy.sparse import hstack
import seaborn as sns
```

```python
df = pd.read_parquet('../data/yellow_tripdata_2023-01.parquet')
df
```

```python
df["duration"] = (df["tpep_dropoff_datetime"] - df['tpep_pickup_datetime']).dt.total_seconds()/60
```

```python
df["duration"].std()
```

```python
(df["duration"] > 60).mean()
```

```python
df2 = df[(df["duration"] <  60) & (df["duration"] > 1)]
len(df2)/len(df)
```

```python
plt.hist(df2.duration, bins=50)
plt.plot()
```

```python
print(df2[["DOLocationID", "PULocationID"]].dtypes)
```

```python
train_dic = df2[["DOLocationID", "PULocationID"]].astype(str).to_dict(orient="records")
dv = DictVectorizer() ## generates sparse matrix
train = dv.fit_transform(train_dic)
```

```python

tr = hstack([train, df2["trip_distance"].values.reshape(-1,1)])
ytr = df2["duration"]
```

```python
tr.shape
```

```python
lr = LinearRegression()
lr.fit(tr, ytr )
yp = lr.predict(tr)
print(np.mean((yp - ytr)**2)**0.5)
```
